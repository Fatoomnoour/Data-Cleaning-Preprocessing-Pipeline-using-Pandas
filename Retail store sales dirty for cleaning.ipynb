```python
# معرفة عدد القيم المفقودة في كل عمود
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# 1. تعويض اسم المنتج المفقود بكلمة Unknown
# Item                1213


df['Item'] = df['Item'].fillna('Unknown')
```

```python
# 2. تعويض القيم العددية (السعر والكمية) بالوسيط Median
# Price Per Unit       609


df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Price Per Unit'].median())
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
```

```python
# 3. معالجة Total Spent (هندسياً: يفضل إعادة حسابه بدلاً من مجرد التعويض)
# Quantity             604

df['Total Spent'] = df['Total Spent'].fillna(df['Price Per Unit'] * df['Quantity'])
```

```python
# 4. تعويض الخصم المفقود بـ 0 (لأن عدم وجود قيمة يعني غالباً عدم تطبيق خصم)
# Total Spent          604

df['Discount Applied'] = df['Discount Applied'].fillna(0)
```

```python

# التأكد من نظافة الداتا تماماً
print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
# حفظ البيانات النظيفة
df.to_csv('retail_store_sales_cleaned.csv', index=False)

print("Data Cleaning Process Completed Successfully!")
```
